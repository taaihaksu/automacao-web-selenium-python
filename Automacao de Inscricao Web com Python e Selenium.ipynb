```python
from selenium import webdriver
# Inicialização do WebDriver para o navegador Chrome
navegador = webdriver.Chrome()
```

```python
# Passo 1: Acesso à URL alvo para o teste de inscrição
navegador.get("https://pages.hashtagtreinamentos.com/inscricao-minicurso-python-automacao-org?origemurl=hashtag_yt_org_minipython_8X_f5p9XUoE")

# Passo 2: Interação com o botão de abertura do modal (Popup)
navegador.find_element('xpath',
                       '//*[@id="BotaoPopup1"]').click()

# Passo 3: Input de dados nos campos do formulário
navegador.find_element('xpath',
                       '//*[@id="firstname"]').send_keys("Gabriel")
navegador.find_element('xpath',
                       '//*[@id="email"]').send_keys("gabriel@outlook.com")
navegador.find_element('xpath',
                       '//*[@id="phone"]').send_keys("11999999999")

# Passo 4: Submissão do formulário via clique no botão de envio
navegador.find_element('xpath',
                       '//*[@id="_form_1925_submit"]').click()

```
